```python
"""

    Binning is nothing we are grouping
    datasets into several chunks.
    
    Binning is useful for data pre-processing.
    
"""
import pandas as pd

inFile = 'files/finalData.csv'
df = pd.read_csv(inFile)

binwidth = (df["price"].max() - df["price"].min())//4
bins = range(df["price"].min(), df["price"].max(), binwidth)
group_name = ['low','medium','high']

df['price-binned'] = pd.qcut(df["price"], 3, labels=group_name)

"""
    One hot enchoding or converting string data into
    some numerical values.
    
"""
data = pd.get_dummies(df["fuel-type"])
df['oneHot_Diesel'] = data['diesel']
df['oneHot_Gas'] = data['gas']
df.to_csv('files/cleanedData.csv')
df.head()
# print(pd.get_dummies(df["fuel-type"]))

```
